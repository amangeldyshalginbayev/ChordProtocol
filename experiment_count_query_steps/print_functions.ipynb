```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1.axes_divider import make_axes_locatable
```

```python
results = np.random.rand(10, 20)
```

```python
im = plt.imshow(results, aspect = 'equal', origin = 'lower', interpolation='bilinear', cmap='RdGy')
ax = plt.gca()
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size='5%', pad=0.5)
plt.colorbar(im, cax=cax, ax=ax)
ax.set_xlabel("Number of keys")
ax.set_xticks([x for x in range(20)])
ax.set_xticklabels(["$2^{"+ str(x) + "}$" for x in range(1, 21)])
ax.set_yticks([x for x in range(10)])
ax.set_yticklabels(["$2^{"+ str(x) + "}$" for x in range(1, 11)])
ax.set_ylabel("Number of nodes")
```
